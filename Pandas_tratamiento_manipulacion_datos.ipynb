# 1. Entendiendo el problema

```python
import pandas as pd
```

```python
datos = pd.read_json('/content/datos_hosting.json')
datos.head()

# trae un diccionario con otros diccionarios adentro
# se llama datos anidados con varios niveles de anidamiento
```

```python
# debemos extraer los diccionarios a columnas

datos = pd.json_normalize(datos['info_inmuebles'])
datos.head()

# hemos transformado un json en un dataframe
```

# 2. Datos numericos

```python
# tenemos variables y listas de numeros y texto, esa información la debemos separar para que se visualice
# vamos a transformar un registro en varios registros

# mostrar los nombres de las columnas
columnas = list(datos.columns)
columnas
```

```python
# vamos a explotar los datos
datos = datos.explode(columnas[3:])
datos

# ya no tenemos listas, pero si diccionarios
```

```python
# Hay indices repetido, debemos recetiearlos

datos.reset_index(drop=True, inplace=True)
datos.head()
```

```python
datos.info()

# no es correcto que el tipo de datos sea obje, debemos corregir esto
```

```python
import numpy as np
```

```python
# vamos a transformar primero los datos enteros

datos['max_hospedes'] = datos['max_hospedes'].astype(np.int64)
datos['max_hospedes']
```

```python
datos.info()
```

```python
# como hay varias columnas numericas vamos a crear una lista para cambiarlas todas al mismo tiempo

col_numerico = ['cantidad_baños', 'cantidad_cuartos', 'cantidad_camas']
```

```python
datos[col_numerico] = datos[col_numerico].astype(np.int64)
```

```python
datos.info()
```

```python
datos['evaluacion_general'] = datos['evaluacion_general'].astype(np.float64)
```

```python
datos.info()
```

```python
# con variables float, pero con caracteres
# Hay que remover el simbolo

datos['precio'] = datos['precio'].apply(lambda x: x.replace('$', '').replace(',','').strip())
# lambda recorre cada valor de esta columna y reemplaza $ por nada'', reemplaza , por nada'', strip quita los espacios para dejar solo el numero
```

```python
datos['precio'] = datos['precio'].astype(np.float64)
```

```python
datos.info()
```

```python
# para cambiar dos coumnas o mas a la vez, debemos tener en cuenta que ya no es una serie sino un df, y apply solo funciona para series

datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].applymap(lambda x: x.replace('$', '').replace(',','').strip())
```

```python
datos[['cuota_deposito', 'cuota_limpieza']] = datos[['cuota_deposito', 'cuota_limpieza']].astype(np.float64)
```

```python
datos.info()
```

Tokenización = es la división del texto en palabras simples para darles valor, ejemplo: excelente, bueno, malo, pesimo. con etos calificativos, se puede variar los presios de un servicio

```python
datos['descripcion_local'] = datos['descripcion_local'].str.lower()
```

```python
  datos.head()
```

```python
# vamos a eliminar los caracteres especiales que no aportan valor en los analisis

datos['descripcion_local'][3169]
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.replace('[^a-zA-Z0-9\-\']',' ', regex=True)
```

```python
datos['descripcion_local'] = datos['descripcion_local'].str.replace('(?<!\w) - (?<!\w)',' ', regex=True)
```
