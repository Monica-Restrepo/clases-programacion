##  **Demandas de ML**

¿Cuáles son los valores promedio de alquiler por tipo de propiedad?

```python
# sacar el promedio en general

datos['Valor'].mean()
```

```python
# pero no esta pidiendo el valor de cada tipo de propiedad, los debemos agrupar por tipo de dato

datos.groupby('Tipo').mean(numeric_only=True)

# el probelma es que saca el promedio de todo, no solo del valor del alquiler
```

```python
datos.groupby('Tipo')['Valor'].mean()

# aca si puedo ver el promedio del valor del alquiler
```

```python
# para ver los valores de menor a mayor

datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
# para hacer mas facil la visualización almacenamos los datos en una variable y así los podemos graficas

df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind='barh', figsize=(10,5), color='skyblue')
```

```python
# vamos a separar las prpiedades residenciales de las industriales

# primero conocer el nombre de los tipos de propiedad

datos.Tipo.unique()
```

```python
# crear una lista con los inmuebles comerciales

inmuebles_comerciales = ['Conjunto Comercial/Sala', 'Edificio Completo', 'Tienda/Salón', 'Casa Comercial', 'Terreno Estándar', 'Cochera/Estacionamiento',
       'Galpón/Depósito/Almacén', 'Tienda en Centro Comercial','Hotel', 'Loteo/Condominio','Industria']
```

```python
datos.query('@inmuebles_comerciales in Tipo')

# Así visualizo la cantidad de inmuebles comerciales en la base de datos
```

```python
# pero lo que queremos es que los inmuebles comerciales no esten

datos.query('@inmuebles_comerciales not in Tipo')
```

```python
df = datos.query('@inmuebles_comerciales not in Tipo')
df.head()
```

```python
# ahora vamos a calcular y graficar el valor promedio de alquiler de las propiedades residenciales

df_tipo_precio = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_tipo_precio.plot(kind='barh', figsize=(10,5), color='purple')
```
