# 1. desafío del proyecto 1: ventas online

```python
import pandas as pd
import requests

# URL del JSON
url = 'https://cdn3.gnarususercontent.com.br/2928-transformacao-manipulacao-dados/dados_vendas_clientes.json'

# Hacer la solicitud HTTP
response = requests.get(url)

# Convertir a JSON
data = response.json()

# Convertir en DataFrame
datos = pd.DataFrame(data)
datos.head()
```

```python
datos = pd.json_normalize(datos['dados_vendas'])
datos.head()
```

## Eliminar datos en listas dentro del DataFrame

```python
columnas = list(datos.columns)
columnas
```

```python
datos = datos.explode(columnas[1:])
datos
```

```python
datos.reset_index(drop=True, inplace=True)
datos.head()
```

```python
datos.info()
```

```python
import numpy as np
```

```python
datos['Valor da compra'] = datos['Valor da compra'].apply(lambda x: x.replace('$', '').replace(',','').replace('R', '').strip())
datos['Valor da compra'] = datos['Valor da compra'].astype(np.float64)
datos.head()
```

```python
datos.info()
```
