# Leyendo banco de datos

## Creando un banco de datos local

```python
import sqlalchemy
```

```python
# hay que importar unos metodos de esa biblioteca
# create_engine = va a crear el motor de datos que va a ejecutar el SQL en nuestro banco de datos
# Metadata que va a guardar la información de las tablas que vamos a tener en nuestro banco de datos
# table para poder trabajar con las tablas
# Inspect para inspeccionar bancos de datos, para saber qué tablas tengo dentro de un banco de datos o inspeccionar una tabla para saber qué atributos tengo dentro de esa tabla.
# text, para convertir las queries en un tipo de datos de texto que se puede entender dentro de las consultas

from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
# hay que crear el motor que quiero usar y donde se va usar

engine = create_engine('sqlite:///:memory:')
```

## Escribiendo un banco de datos

```python
import pandas as pd
```

```python
archivo = 'clientes_banco.csv'
```

```python
datos = pd.read_csv(archivo)
```

```python
datos.head()
```

```python
# debemos exportarlo a formato sql

datos.to_sql('clientes', engine, index=False)
```

```python
  inspector = inspect(engine)
```

```python
# para mirar que tablas contine

print(inspector.get_table_names())
```

## leyendo una consulta SQL

```python
# select *: para seleccionar todas las columnas
# from: el nombre de la tabla
# where: la condición que le doy a la consulta

query = 'SELECT * FROM clientes WHERE Categoria_de_renta	= "Empleado"'
```

```python
empleados =  pd.read_sql(sql = text(query), con = engine.connect())
```

```python
empleados
```

```python
empleados.to_sql('empleados', con = engine.connect(), index=False,if_exists='replace')
```

```python
# esta forma es para consultar la tabla completa

pd.read_sql_table('empleados', con = engine.connect())
```

```python
# para consultar algunas columnas

pd.read_sql_table('empleados', con = engine.connect(), columns=['ID_Cliente', 'Grado_estudio', 'Rendimiento_anual'])
```

## Actualizando un banco de  datos

```python
query = 'SELECT * FROM clientes'
```

```python
pd.read_sql(sql=text(query), con=engine.connect())
```

```python
# vamos a borrar un registro
# debenos importar la siguiente biblioteca para que nos ayude a identificar errores en caso de que se presenten

from sqlalchemy.exc import SQLAlchemyError

query = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
try:
  r_set = engine.connect().execute(text(query))
except SQLAlchemyError as e:
    print(e)
else:
    print('#Registros borrados: ', r_set.rowcount)

    # me muestra si la ejecución tuvo algún probela
```

```python
# confirmar que borro solo un registro

pd.read_sql_table('clientes', con= engine.connect())
```

```python
# vamos a actulizar un registro

query = 'UPDATE clientes SET Grado_estudio="Nivel superior" WHERE ID_Cliente=5008808'
try:
  r_set = engine.connect().execute(text(query))
except SQLAlchemyError as e:
    print(e)
else:
    print('#Registros actualizados: ', r_set.rowcount)
```

```python
pd.read_sql_table('clientes', con= engine.connect())
```

```python
inspector = inspect(engine)
print(inspector.get_table_names())
```
